```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/My Drive/twitter_2009.csv', encoding='latin-1')
df
```

```python
df.columns = ['A', 'B', 'Date', 'C', 'UserName', 'Text']
df
```

```python
#这是使用语法,模糊匹配
df_1 = df[df['Text'].str.contains("i feel|i am feeling|i'm feeling|i dont feel|I'm|Im|I am|makes me")]
df_1 = df_1[~df_1['Text'].str.contains("http:|www")]
df_1.reset_index(drop=True, inplace=True)
df_1
```

```python
df_1['Weekday'] = df_1['Date']

for i in range(len(df_1['Date'])):
  df_1['Weekday'][i] = df_1['Date'][i][:3]
```

```python
df_1 = df_1.drop(df_1[(df_1['Weekday']=='Sat') | (df_1['Weekday']=='Sun')].index)
df_1
```

```python
df_1['DATE'] = df_1['Date']
df_1['Date'] = pd.to_datetime(df_1['DATE']).dt.date
df_1['Date'] = pd.to_datetime(df_1['Date']).dt.strftime('%Y-%m-%d')
df_1.reset_index(drop=True, inplace=True)
df_1
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# Sentiment Analysis
analyzer = SentimentIntensityAnalyzer()

scores = df_1['Text'].apply(analyzer.polarity_scores).tolist()

df_scores = pd.DataFrame(scores)
df_2 = df_1.join(df_scores, rsuffix='_right')
```

```python
df_2
```

```python
unique_date = df_2['Date'].unique().tolist()
text_dict = {date: df_2.loc[df_2['Date'] == date] for date in unique_date}
text_dict
```

```python
values_compund = []
values_neg = []
values_neu = []
values_pos = []

for date in unique_date: 
    dataframe = text_dict[date]
    dataframe = dataframe.drop(columns = ['Text'])
    mean_compound = round(dataframe['compound'].mean(), 2)
    mean_neg = round(dataframe['neg'].mean(), 2)
    mean_neu = round(dataframe['neu'].mean(), 2)
    mean_pos = round(dataframe['pos'].mean(), 2)
    values_compund.append(mean_compound)
    values_neg.append(mean_neg)
    values_neu.append(mean_neu)
    values_pos.append(mean_pos)
    
df_3 = pd.DataFrame(list(zip(unique_date, values_compund, values_neg, values_neu, values_pos)), columns =['Date', 'Mean Compund', 'Mean Negtive', 'Mean Nature', 'Mean Positive']) 
#df_3 = df_3.set_index('Date')
#df_3 = df_3.sort_values('Mean Sentiment', ascending=False)
df_3
```

```python
df_3 = df_3.drop(index=[13])
df_3
```

```python
import numpy as np
# k值可改变
k = 4

def z_score(x, k):
  list = []
  for i in range(len(x)):
    mean = np.mean(x.iloc[i-k : i+k+1])
    std = np.std(x.iloc[i-k : i+k+1])
    z = round((x.iloc[i]-mean)/std, 2)
    list.append(z)
  return list

df_k = df_3
df_k['Mean Compund'] = z_score(df_k['Mean Compund'], k)
df_k['Mean Negtive'] = z_score(df_k['Mean Negtive'], k)
df_k['Mean Nature'] = z_score(df_k['Mean Nature'], k)
df_k['Mean Positive'] = z_score(df_k['Mean Positive'], k)
df_k['Change'] = z_score(df_k['Change'], k)
df_k
```
